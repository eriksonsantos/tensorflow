```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import tensorflow as tf
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([-7, -4, -1 , 2, 5, 8, 11 , 14])

y = np.array([3, 6, 9, 12, 15, 18, 21, 24])

plt.scatter(X,y)
```

```python
y == X + 10
```

```python
house_info = tf.constant(['bedroom', 'bathroom', 'garage'])
house_price = tf.constant([939700])

house_info, house_price
```

```python
X = tf.cast(tf.constant(X), dtype=tf.float32)
y = tf.cast(tf.constant(y), dtype=tf.float32)
```

```python
plt.scatter(X,y)
```

```python
tf.random.set_seed(42)

#Create a model using the Sequential API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
    
])

#Compile the model

model.compile(loss= tf.keras.losses.mae, #Mae representative Mean Absolute error
              optimizer=tf.keras.optimizers.SGD(), #Sgd representative Stochastic gradient descent
              metrics=["mae"]
             )

model
```

```python
#Fitting the model
model.fit(tf.expand_dims(X, axis=-1),y, epochs=100);
```

```python
#Improve the model

tf.random.set_seed(42)

#Create a model using the Sequential API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(50, activation="relu"),
    tf.keras.layers.Dense(1)
    
])

#Compile the model

model.compile(loss= tf.keras.losses.mae, #Mae representative Mean Absolute error
              optimizer=tf.keras.optimizers.Adam(lr =0.01), #Sgd representative Stochastic gradient descent
              metrics=["mae"]
             )

model.fit(tf.expand_dims(X, axis=-1),y, epochs=1000);
```

```python
def plot_predictions(train_data=X_train,
                    train_labels=y_train,
                    test_data= X_test,
                    test_labels = y_test,
                    predictions=y_pred):
    
    plt.figure(figsize=(10,7))
    plt.scatter(train_data, train_labels, c="b", label="Training data")
    plt.scatter(test_data, test_labels, c="r", label="Testing data")
    
    plt.scatter(test_data, predictions, c="g", label="Predictions")
    
    plt.legend()
```
